```python
import pandas as pd

long_trades= pd.read_parquet('data/long_renko_trades.parquet',engine='pyarrow')
long_trades.head(20)
```

```python
short_trades = pd.read_parquet( 'data/short_renko_trades.parquet', engine = 'pyarrow')
short_trades
```

```python
#Add addition columns to destinguish long and short trades

long_trades['trade_type'] = 'long'
short_trades['trade_type'] = 'short'
```

```python
combined_df = pd.concat([long_trades, short_trades], ignore_index=True)
combined_df.head(10)
```

```python
#one-hot group identification

columns_to_encode = [
    'strategy', 'neutral_gex', 'gamma_flip_strike_low_vs_hist', 
    'gamma_flip_strike_high_vs_hist', 'gamma_flip_strike_not_near', 
    'max_call_wall_strike_low_vs_hist', 'max_call_wall_strike_high_vs_hist', 
    'max_call_wall_strike_not_near', 'min_put_wall_strike_low_vs_hist', 
    'min_put_wall_strike_high_vs_hist', 'min_put_wall_strike_not_near', 
    'neutral_zone', 'call_wall_compression', 'put_wall_compression', 
    'short_gamma_expansion', 'gamma_valley', 'wall_to_wall_compression', 
    'trade_type'
]

# prefix dictionary
prefix_dict = {col: col for col in columns_to_encode}


combined_df_encoded = pd.get_dummies(
    combined_df,
    columns=columns_to_encode,
    prefix=prefix_dict,
    prefix_sep='_'
)

print(f"Encoded columns: {list(combined_df_encoded.columns)}")
```

```python
combined_df_encoded
```

```python
# Remove '_Fasle ' Features to reduce redundancy

all_cols = combined_df_encoded.columns.tolist()

true_cols = [col for col in all_cols if col.endswith('_True')]
false_cols = [col for col in all_cols if col.endswith('_False')]

non_bool_cols = [col for col in all_cols if not (col.endswith('_True') or col.endswith('_False'))]

combined_df_encoded = combined_df_encoded[non_bool_cols + true_cols]

print(f"Dropped redundant False columns: {false_cols}")
print(f"Remaining columns: {list(combined_df_encoded.columns)}")
```

```python
# updating features_to_drop.txt
with open('features_to_drop.txt', 'w') as f:
    f.write("REDUNDANT FEATURES FOR REMOVAL\n")
    
    f.write("Redundant False Columns (Boolean Pairs)\n")
    f.write("-"*60 + "\n")
    f.write(f"Count: {len(false_cols)}\n")
    for col in sorted(false_cols):
        f.write(f"  • {col}\n")
    f.write("\n")
```

```python
# Variance Detection

from sklearn.feature_selection import VarianceThreshold

# numeri columns only
numeric_df = combined_df_encoded.select_dtypes(include=['number'])

# variance threshold
threshold = 0.01  
selector = VarianceThreshold(threshold=threshold)
selector.fit(numeric_df)

# Get indices of features 
features_to_keep = numeric_df.columns[selector.get_support()].tolist()
features_dropped = numeric_df.columns[~selector.get_support()].tolist()

dropped_variances = numeric_df[features_dropped].var().sort_values()
kept_variances = numeric_df[features_to_keep].var().sort_values()

kept_features = {col: float(numeric_df[col].var()) for col in features_to_keep}

print(f"VARIANCE THRESHOLD: {threshold}")
print(f"Total numeric features: {len(numeric_df.columns)}")
print(f"Features kept: {len(features_to_keep)}")
print(f"Features dropped: {len(features_dropped)}\n")

print("DROPPED FEATURES (with variance values):")
print("=======")
for feature in features_dropped:
    variance = numeric_df[feature].var()
    print(f"  {feature}: {variance:.6f}")

print("\n" + "=======")
print("KEPT FEATURES:")

for feature in features_to_keep:
    variance = numeric_df[feature].var()
    print(f"  {feature}: {variance:.6f}")

# Keep both the selected numeric features + any non-numeric columns you want
string_cols = ['name', 'symbol']  # Or whatever your string columns are
combined_df_new = combined_df_encoded[features_to_keep + string_cols]
```

```python
# updating features_to_drop.txt
with open('features_to_drop.txt', 'a') as f:
    f.write("REDUNDANT FEATURES FOR REMOVAL\n")
    
    f.write("Redundant Features (Low Variance)\n")
    f.write("-"*60 + "\n")
    f.write(f"Count: {len(features_dropped)}\n")
    for col in sorted(features_dropped):
        f.write(f"  • {col}\n")
    f.write("\n")
```

```python
# Now do correlation analysis on combined_df
correlation_matrix = combined_df_new.corr(method='spearman')
```

```python
# Find highly correlated pairs
def find_redundant_features(corr_matrix, threshold=0.9):
    redundant = []
    for i in range(len(corr_matrix.columns)):
        for j in range(i+1, len(corr_matrix.columns)):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                col1 = corr_matrix.columns[i]
                col2 = corr_matrix.columns[j]
                redundant.append((col1, col2, corr_matrix.iloc[i, j]))
    return redundant

redundant_pairs = find_redundant_features(correlation_matrix, threshold=0.9)
for feature1, feature2, corr in redundant_pairs:
    print(f"{feature1} <-> {feature2}: {corr:.3f}")
```

```python
kept_features
```

```python
# drop one feature from each pair = lowest variance feature 

features_to_drop_corr = []
drop_info = []

for feat1, feat2, corr_value in redundant_pairs:
    # Get variance for both features
    var1 = kept_features.get(feat1, None)
    var2 = kept_features.get(feat2, None)
    
    if var1 is None or var2 is None:
        print("Warning: Feature not found in kept_features dict")
        continue
    
    # Drop the one with lower variance
    if var1 < var2:
        to_drop = feat1
        to_keep = feat2
    else:
        to_drop = feat2
        to_keep = feat1
    
    features_to_drop_corr.append(to_drop)
    drop_info.append({
        'dropped': to_drop,
        'kept': to_keep,
        'correlation': corr_value,
        'dropped_variance': min(var1, var2),
        'kept_variance': max(var1, var2)
    })
    
    print(f"Pair: {feat1} <-> {feat2}")
    print(f"  Correlations: {corr_value:.4f}")
    print(f"  {feat1} variance: {var1:.6f}")
    print(f"  {feat2} variance: {var2:.6f}")
    print(f"  → Dropping: {to_drop} (variance: {min(var1, var2):.6f})")
    print()

print("=====")
print(f"Total features to drop: {len(features_to_drop_corr)}")
print(f"Features: {features_to_drop_corr}\n")

# Drop features
combined_df_final = combined_df_new.drop(columns=features_to_drop_corr)
print(f"Dropped {len(features_to_drop_corr)} low-variance correlated features")
print(f"Final shape: {combined_df_final.shape}")

```

```python
# updating features_to_drop.txt
with open('features_to_drop.txt', 'a') as f:
    f.write("REDUNDANT FEATURES FOR REMOVAL\n")
    
    f.write("Redundant False Features (High Correlation Pairs ; lowest variance)\n")
    f.write("-"*60 + "\n")
    f.write(f"Count: {len(features_to_drop_corr)}\n")
    for col in sorted(features_to_drop_corr):
        f.write(f"  • {col}\n")
    f.write("\n")
```
